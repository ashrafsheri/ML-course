```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3,4,5,6])
Y = np.array([1,2,3,4,5,6])
```

```python
y_func = 0.5*X + 0.5
```

```python
plt.scatter(X,Y)
plt.plot(X,y_func)
plt.show()  
```

```python
def cost_function(x,y,w,b):
    m = X.shape[0]  
    cost = 0
    for i in range(m):
        f_wb = w*X[i] + b
        cost += (f_wb - Y[i])**2
    total_cost = cost/(2*m)
    return total_cost
```

```python
def compute_gradient(x,y,w,b,learning_rate):
    m = X.shape[0]
    dw = 0
    db = 0
    for i in range(m):
        f_wb = w * X[i] + b
        dw_i = (f_wb - Y[i])*X[i]
        db_i = (f_wb - Y[i])
        dw += dw_i
        db += db_i
    dw = dw/m
    db = db/m
    return dw, db
```

```python
def gradient_descent(x,y,w,b,learning_rate,epochs):
    cost_history = []
    p_history = []
    for i in range(epochs):
        dw, db = compute_gradient(x,y,w,b,learning_rate)
        w = w - learning_rate*dw
        b = b - learning_rate*db
        p_history.append([w,b])
        cost = cost_function(x,y,w,b)
        cost_history.append(cost)
    return w, b, cost_history, p_history
```

```python
w,b,cost_history,p_history = gradient_descent(X,Y,0,0,0.00001,100000000)
# print(f'w:{data[0]}\n b:{data[1]}\n Cost:{data[2][-1]}')
```

```python
w_values = [pair[0] for pair in p_history]
plt.plot(w_values,cost_history)
plt.show()
```
